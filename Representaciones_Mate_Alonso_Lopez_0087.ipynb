#### Problema 1
La suma de un número más nueve es igual a 21Paso 1: Análisis y SoluciónPlanteamiento: Definimos $x$ como el número que queremos encontrar. La frase "la suma de un número más el número nueve" se escribe como $x + 9$. El enunciado dice que esto "es igual a 21".Ecuación:$$x + 9 = 21$$Despeje: Para dejar a la variable $x$ sola, el 9 que está sumando pasa al otro lado de la igualdad con la operación contraria (resta):$$x = 21 - 9$$Solución:$$x = 12$$

```python
# Problema 1: Hallar un número x tal que x + 9 = 21

# Definimos el resultado total y el valor conocido
# total = 21
# sumando_conocido = 9
resultado_total = int(input("Ingrese el resultado total: "))
sumando_conocido = int(input("Ingrese el sumando conocido: "))
# Despejamos x: x = total - sumando_conocido
x = resultado_total - sumando_conocido

print(f"El número buscado es: {x}")
```

##### Problema 2
Edades de Pedro y AnaPaso 1: Análisis y SoluciónPlanteamiento:Edad de Ana: La llamaremos $x$ (es nuestra referencia).Edad de Pedro: El problema dice que es "el doble de edad que su hermana", por lo tanto es $2x$.La suma de ambas edades es 36.Ecuación:$$x + 2x = 36$$Simplificación: Sumamos los términos semejantes ($1x + 2x$):$$3x = 36$$Despeje: El 3 que está multiplicando a la $x$ pasa al otro lado dividiendo:$$x = \frac{36}{3}$$Solución:$x = 12$ (Edad de Ana)$2x = 2(12) = 24$ (Edad de Pedro)Comprobación: $12 + 24 = 36$.

```python
# Problema 2: Pedro (2x) y Ana (x) suman 36 años

# suma_edades = 36
suma_edades = int(input("Ingrese la suma de las edades: "))
# La ecuación es 3x = 36, por lo tanto x = 36 / 3
edad_ana = suma_edades / 3
edad_pedro = edad_ana * 2

print(f"La edad de Ana es: {int(edad_ana)} años")
print(f"La edad de Pedro es: {int(edad_pedro)} años")
```
